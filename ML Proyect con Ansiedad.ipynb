En este notebook probaremos conservando la ansiedad

Codificamos nuestras variables categóricas

```python
df['gender'] = df['gender'].map({1: 'Male', 2: 'Female', 3: 'Other'})
df['orientation'] = df['orientation'].map({1: 'Heterosexual', 2: 'Bisexual', 3: 'Homosexual', 4: 'Asexual', 5: 'Other'})
```

```python
df.head(15)
```

```python
df.info()
```

Observamos que tenemos unos pocos nulos en sexo y orientación sexual, siendo categoras tan sensibles, creemos que en este caso lo mejor es eliminar estas filas

```python
df = df.dropna()
```

```python
df = pd.get_dummies(df, drop_first = True)
df
```

A partir de ahora vamos a realizar acciones que podrian contaminar nuestros datos, por lo tanto lo primero que vamos a hacer es dividir en train y test

```python
X = df.drop('Depression_total', axis = 1)
y = df['Depression_total']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

Las puntuaciones de esta base de datos estan muy por encima de lo que deberian ser, ya que cuando cuando la puntuación 
en depresion es mayor de 28 ya se considera una caso extremadamente severo. Seguramente se debe a que la gran mayoria de sujetos 
son de otra cultura, asiatica en este caso. 

Ya que es muy complicado encontrar bases de datos fiables, con los test realizados por la misma persona, voy a usar la mediana como marcador
de si el sujeto deberia ser atendido por depresión o no, a modo de experimento. 

Más adelante se podria repetir, como hemos comentado anteriormente, con datos obtenidos de forma autonoma o facilitados por alguna empresa o colegio/instituto interesados

```python
print('Media depresión: ', y_train.mean())
print('Mediana depresión: ', y_train.median())
print('Media ansiedad: ', X_train['Anxiety_total'].mean())
print('Mediana ansiedad: ', X_train['Anxiety_total'].median())
print('Media stress: ', X_train['Stress_total'].mean())
print('Mediana stress: ', X_train['Stress_total'].median())
```

Codificamos la depresión como hemos hablado en train y test usando el valor de la mediana de TRAIN

```python
y_train= y_train.apply(lambda x:1 if x > 35 else 0)
```

```python
y_test= y_test.apply(lambda x:1 if x > 35 else 0)
```

Escalamos los datos

```python
scaler = MinMaxScaler()
numerical = ['age', 'Anxiety_total', 'Stress_total']

scaler.fit(X_train[numerical])
X_train[numerical] = scaler.transform(X_train[numerical])
X_test[numerical] = scaler.transform(X_test[numerical])
```

```python
X_train.head()
```

Empezamos la Regresión Logística

```python
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
name = 'Logistic Regression'
kfold = model_selection.KFold(n_splits = 10)
cv_results = model_selection.cross_val_score(model, X_train, y_train, cv = kfold, scoring = 'recall')
msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(msg)
```

```python
predictions = model.predict(X_test)
print(accuracy_score(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```
